```python
%matplotlib inline
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
from IPython.display import set_matplotlib_formats
from filter import *

# set_matplotlib_formats('svg')
pd.set_option('display.max_rows', None)
cut_ratio = 0.05
```

```python
print("\n##### UPLINK MESSAGES #####\n")
file = "uplink_messages.csv"
data = pd.read_csv(file, delimiter=',')

### FILTERING ###
enable_filtering = True

config = {
    'node_id': [],
    'ap_id': [],
    'snr': [],
    'rssi': [],
    'spf': [],
    'frequency': [],
    'app_data': [],
    'recv_time_from': "",
    'recv_time_to': ""
}
operator = "AND"
### FILTERING ###

filtered_data = filter(data, config, enable_filtering, operator)

del filtered_data['id']
del filtered_data['msg_group_number']
del filtered_data['is_primary']
del filtered_data['message_type_id']
del filtered_data['coderate']
del filtered_data['bandwidth']
del filtered_data['duty_cycle_remaining']
del filtered_data['seq']
del filtered_data['power']
del filtered_data['airtime']
del filtered_data['application_id']
del filtered_data['battery']
del filtered_data['dev_id']
                
sorted_data = filtered_data.sort_values(by='receive_time', ascending=True)
display(sorted_data)
```
